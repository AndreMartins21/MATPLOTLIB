# GRÁFICOS SIMPLES

- plt.plot()

```
import matplotlib.pyplot as plt
import matplotlib as mpl

# GRÁFICO SIMPLES <plt.plot()>

plt.plot([1, 2, 3, 4], [2, 4, 8, 16])
plt.show()
```

```
# GRÁFICO MELHORADO

indice = list(range(10))

base = list(map(lambda x : 2**x, indice))

plt.plot(indice, base)
plt.xlabel('Indices')
plt.ylabel('Potência')
plt.title('Potência de 2')
plt.show()
```

# GRÁFICO DE BARRAS 

- plt.bar()

```
# IMPORTAÇÕES

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```
# 1° EXEMPLO

indice = list(range(10))

base = list(map(lambda x : 2**x, indice))

plt.bar(indice, base, label='Potenciação', color = 'g')
plt.legend()
plt.show()
```

```
# 2° EXEMPLO
x = [0, 4, 11, 12, 19]

y = [1, 5, 10, 13, 18]

x2 = [2, 6, 9, 14, 17]

y2 = [3, 7, 8, 15, 16]

plt.bar(x, y, label='Menores', color='b')
plt.bar(x2, y2, label='Maiores', color='r')
plt.legend()
plt.show()
```

```
# 3° EXEMPLO

idade = [2, 29, 23, 1, 9, 32, 12, 4, 19, 17, 38]

ids = [x for x in range(len(idade))]

plt.bar(idade, ids, label='Idades', color='g')
plt.legend()
plt.show()
```

# HISTOGRAMA

- plt.hist()

```
import matplotlib.pyplot as plt
import matplotlib as mpl


# EM FORMA DE BARRAS <histtype='bar'>

idade = [5, 13, 19, 29, 35, 31, 43]

otter_age = [0, 10, 20, 30, 40, 50, 60]

plt.hist(idade, otter_age, histtype='bar', rwidth=0.5)
plt.title('Pessoas por faixa etária')
plt.show()
```

```
# EM FORMA DE STEPFILLED <histtype='stepfilled'>

idade = [5, 13, 19, 29, 35, 31, 43]

otter_age = [0, 10, 20, 30, 40, 50, 60]

plt.hist(idade, otter_age, histtype='stepfilled', rwidth=0.5)
plt.title('Pessoas por faixa etária')
plt.show()
```

# SCATTERPLOT

- Marcação de pontos coordenadas;

- plt.scatter()

```
import matplotlib.pyplot as plt

x_impar = list(range(1, 20, 2))

y_par = list(range(0, 19, 2))

plt.scatter(x_impar, y_par, label = 'Pontos (PAR/IMPAR)', color = 'r', s=50)
plt.scatter(np.arange(10), np.arange(10, 20), label = 'Sequência', color = 'g', s=70)
plt.legend(title='< Pontos no gráfico >')
plt.show()
```

# STACK PLOTS

- Mostrar área empilhada em um gráfico;

- plt.stackplot()

```
import matplotlib.pyplot as plt
import numpy as np


a = list(range(5))
b = list(range(5, 10))
c = list(range(10, 15))
d = list(range(15, 20))

plt.stackplot(a, b, c, d, labels=['A', 'B', 'C', 'D'], colors = ['g', 'b', 'y', 'c'])
plt.legend(title='#Legenda:')
plt.show()
```

# PIE CHART

- Gráfico de Pizza;

- plt.pie()

```
import matplotlib.pyplot as plt

fatias = [63, 37]
rotulo = ['Homem', 'Mulher']
coluna = ['b', 'm']

plt.pie(fatias, labels = rotulo, colors= coluna, shadow=True, startangle=30, explode= (0, 0.09))
plt.title('# Participação em graduação por sexo')
plt.show()
```

# SUBPLOTS

- Criar vários plots no mesmo local para comparação;

- var, var_2 = plt.subplots(SHAPE)

```
import matplotlib.pyplot as plt
from random import choice


# CRIANDO SUBPLOTS <tupla, var = plt.subplots(SHAPE)>

sub, ax = plt.subplots(2, 2, figsize = (9, 6))

cores = ['red', 'blue', 'green', 'yellow', 'black', 'pink']

for y in range(2):
  for x in range(2):
    ax[y, x].bar(np.random.randn(6), np.random.randn(6), color = choice(cores))


plt.show()
```

```
_, axes = plt.subplots(1, 2, figsize = (10, 10))

a = np.arange(10)
b = np.arange(10, 30, 2)
c = np.random.randn(10)

axes[0].plot(a, b, c)
axes[0].set_title('Teste 1')
axes[0].grid(True)

axes[1].plot(c)
axes[1].set_title('Teste 2')
axes[1].grid(True)

plt.show()
```
